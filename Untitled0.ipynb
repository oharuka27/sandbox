```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 130, where=(ys > 130), facecolor='g', alpha=0.1)

plt.title("Sample Visualization")
plt.show()
```
